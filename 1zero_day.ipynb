```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'preprocessing_ofCIC-IDS2017.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
df[' Label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE=LabelEncoder()
```

```python
df[' Label']=LE.fit_transform(df[' Label'])
```

```python
df.sample()
```

```python
X=df.drop(' Label',axis=1)
y=df[' Label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
ss=StandardScaler()
```

```python
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
LR=LogisticRegression()
DTC=DecisionTreeClassifier()
RFC=RandomForestClassifier()
KNN=KNeighborsClassifier()
```

### Logistic Regrssion

```python
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
from sklearn.metrics import recall_score,classification_report,confusion_matrix,f1_score
```

```python
print('recall:',recall_score(y_test,y_pred,average='micro'))
```

```python
print(classification_report(y_test,y_pred))
```

```python
LR.score(x_test,y_test)
```

### Decision tree

```python
DTC.fit(x_train,y_train)
```

```python
DTC_pred=DTC.predict(x_test)
```

```python
print('Classification Report:','\n',classification_report(y_test,DTC_pred))
print('Recall:',recall_score(y_test,DTC_pred,average='weighted'))
```

### RandomForest


```python
RFC.fit(x_train,y_train)
```

```python
RFC_pred=RFC.predict(x_test)
```

```python
print('Classification Report:','\n',classification_report(y_test,RFC_pred))
print('Recall:',recall_score(y_test,RFC_pred,average='weighted'))
```

### KNN

```python
KNN.fit(x_train,y_train)
```

```python
KNN_pred=KNN.predict(x_test)
```

```python
print('Classification Report:','\n',classification_report(y_test,KNN_pred))
print('Recall:',recall_score(y_test,KNN_pred,average='weighted'))
```
